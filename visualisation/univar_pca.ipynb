# Univariate plots, transformed data

```python
import pandas as pd

iris = pd.read_csv("../data/iris_cleaned.csv")
```

## Full dataset projection

```python
from visualisation import run_pca

x_transformed, _, _ = run_pca(iris)

iris_trans = pd.DataFrame(x_transformed, columns=[f"PC{i+1}" for i in range(x_transformed.shape[1])])
iris_trans["species"] = iris["species"]
```

```python
from visualisation import hist_plot

for var in iris_trans.columns[:-1]:
    hist_plot(iris_trans, var)
```

## Univariate projection

```python
x_transformed, _, _ = run_pca(iris, projection="virginica")

iris_trans = pd.DataFrame(x_transformed, columns=[f"PC{i+1}" for i in range(x_transformed.shape[1])])
iris_trans["species"] = iris["species"]
```

```python
for var in iris_trans.columns[:-1]:
    hist_plot(iris_trans, var)
```
